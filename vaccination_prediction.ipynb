```python
from sklearn.preprocessing import LabelEncoder

le_concern = LabelEncoder()
df['h1n1_concern'] = le_concern.fit_transform(df['h1n1_concern'])

le_knowledge = LabelEncoder()
df['h1n1_knowledge'] = le_knowledge.fit_transform(df['h1n1_knowledge'])

le_b1 = LabelEncoder()
df['behavioral_antiviral_meds'] = le_b1.fit_transform(df['behavioral_antiviral_meds'])

le_b2 = LabelEncoder()
df['behavioral_avoidance'] = le_b2.fit_transform(df['behavioral_avoidance'])

le_b3 = LabelEncoder()
df['behavioral_face_mask'] = le_b3.fit_transform(df['behavioral_face_mask'])

le_b4 = LabelEncoder()
df['behavioral_wash_hands'] = le_b4.fit_transform(df['behavioral_wash_hands'])

le_b5 = LabelEncoder()
df['behavioral_large_gatherings'] = le_b5.fit_transform(df['behavioral_large_gatherings'])

le_b6 = LabelEncoder()
df['behavioral_outside_home'] = le_b6.fit_transform(df['behavioral_outside_home'])

le_b7 = LabelEncoder()
df['behavioral_touch_face'] = le_b7.fit_transform(df['behavioral_touch_face'])

le_dr_h1n1 = LabelEncoder()
df['doctor_recc_h1n1'] = le_dr_h1n1.fit_transform(df['doctor_recc_h1n1'])

le_dr_seas = LabelEncoder()
df['doctor_recc_seasonal'] = le_dr_seas.fit_transform(df['doctor_recc_seasonal'])

le_chronic = LabelEncoder()
df['chronic_med_condition'] = le_chronic.fit_transform(df['chronic_med_condition'])

le_child = LabelEncoder()
df['child_under_6_months'] = le_child.fit_transform(df['child_under_6_months'])

le_hcw = LabelEncoder()
df['health_worker'] = le_hcw.fit_transform(df['health_worker'])

le_ins = LabelEncoder()
df['health_insurance'] = le_ins.fit_transform(df['health_insurance'])
```

```python
le_age = LabelEncoder()
df['age_group'] = le_age.fit_transform(df['age_group'])

le_edu = LabelEncoder()
df['education'] = le_edu.fit_transform(df['education'])

le_race = LabelEncoder()
df['race'] = le_race.fit_transform(df['race'])

le_sex = LabelEncoder()
df['sex'] = le_sex.fit_transform(df['sex'])

le_income = LabelEncoder()
df['income_poverty'] = le_income.fit_transform(df['income_poverty'])

le_marital = LabelEncoder()
df['marital_status'] = le_marital.fit_transform(df['marital_status'])

le_rent = LabelEncoder()
df['rent_or_own']= le_rent.fit_transform(df['rent_or_own'])

le_employ = LabelEncoder()
df['employment_status'] = le_employ.fit_transform(df['employment_status'])
```

```python
X = df.drop(columns='h1n1_vaccine')
y = df['h1n1_vaccine']
```

```python
X.info()
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(random_state=20211122)
clf.fit(X, y.values)
```
